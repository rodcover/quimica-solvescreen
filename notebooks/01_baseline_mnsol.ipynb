1. Obtain **MNSOL** under [UMN license](https://comp.chem.umn.edu/mnsol/) (do not redistribute raw files).
2. Run ETL → Parquet, then train; compare **MAE/RMSE** (kcal/mol) to [Delfos (*Chem. Sci.* 2019)](https://doi.org/10.1039/C9SC02452B) and newer baselines (e.g. [SolvBERT](https://doi.org/10.1039/D2DD00107A)).

From repo root, `PYTHONPATH=src` or use `scripts/` wrappers.

```python
import subprocess
from pathlib import Path

root = Path("..").resolve()
fixture = root / "data" / "fixtures" / "sample_mnsol_like.tsv"
out_pq = root / "data" / "processed" / "notebook_sample.parquet"
subprocess.check_call(
    ["python", str(root / "scripts" / "run_etl.py"), str(fixture), "-o", str(out_pq)],
    cwd=root,
)
print(out_pq.stat().st_size, "bytes")
```

```python
import json
import subprocess
from pathlib import Path

root = Path("..").resolve()
out_pq = root / "data" / "processed" / "notebook_sample.parquet"
model_path = root / "models" / "notebook_baseline.joblib"
metrics_path = root / "models" / "notebook_metrics.json"
subprocess.check_call(
    [
        "python",
        str(root / "scripts" / "run_train.py"),
        str(out_pq),
        "-o",
        str(model_path),
        "--metrics",
        str(metrics_path),
    ],
    cwd=root,
)
print(json.loads(metrics_path.read_text()))
```
